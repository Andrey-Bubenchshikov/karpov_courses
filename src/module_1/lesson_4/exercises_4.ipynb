```python
import pandas as pd 
path_to_file = '3_taxi_peru.csv'
```

```python
types = ['asap', 'reserved']
taxi = pd.read_csv(filepath_or_buffer=path_to_file, sep=';', parse_dates=['start_at', 'end_at', 'arrived_at'])
taxi.query("start_type in @types") #answer
```

## Task 9

```python
taxi = taxi.query("start_type in @types") 
(
    taxi
     .assign(wait_time = lambda x: (x.arrived_at - x.start_at).astype('timedelta64[m]'))
)
```

## Task 10

```python
(
    taxi
     .assign(wait_time = lambda x: (x.arrived_at - x.start_at).astype('timedelta64[m]'))
     .query("start_type == 'reserved' and wait_time > 0.0")
     .groupby('driver_id', as_index=False)
     .agg(journeys = ('journey_id', 'count'))
     .sort_values(by='journeys', ascending=False)
)
```

## Task 15

```python
import matplotlib.pyplot as plt

taxi_mau = (
    taxi
     .assign(
         month = lambda x: x.start_at.dt.month,
         weekday = lambda x: x.start_at.dt.day_name()
     )
     .groupby('month', as_index=False)
     .agg(mau = ('user_id', 'nunique'))
)

sns.lineplot(data = taxi_mau, x='month', y='mau')
sns.despine()
plt.xticks(range(1, 13))
plt.yticks(range(0, 650, 50))
plt.show()
# answer
# 1. MAU за апрель примерно в 2.5 раза больше, чем за январь
# 2. В течение всего года число уникальных пользователей увеличивалось

```
